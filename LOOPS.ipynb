```python
# Initialize menu items and prices
items = ["Coffee", "Tea", "Green Tea"]
prices = [10, 20, 30]

# Initialize shopping cart
cart = []

def show_menu():
    print("╔════════════════════════════╗")
    print("║       MENU AND PRICES     ║")
    print("╟────────────────────────────╢")
    
    for i in range(len(items)):
        print(f"║ {i+1}. {items[i]:<15} ${prices[i]:<5} ║")

    print("╚════════════════════════════╝")

def show_cart():
    print("╔════════════════════════════╗")
    print("║         YOUR CART         ║")
    print("╟────────────────────────────╢")

    total_cost = 0

    for i in range(len(cart)):
        item, price = cart[i]
        total_cost += price
        print(f"║ {i+1}. {item:<15} ${price:<5} ║")

    print("╟────────────────────────────╢")
    print(f"║ Total: ${total_cost:<13} ║")
    print("╚════════════════════════════╝")

def main():
    shopping = True

    while shopping:
        print("\n1. View Menu")
        print("2. Add to Cart")
        print("3. View Cart")
        print("4. Checkout")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            show_menu()
        elif choice == '2':
            show_menu()
            selection = int(input("Select an item to add to your cart: ")) - 1

            if 0 <= selection < len(items):
                cart.append((items[selection], prices[selection]))
                print(f"{items[selection]} has been added to your cart.")
            else:
                print("Invalid selection. Please choose a valid item.")

        elif choice == '3':
            show_cart()

        elif choice == '4':
            if len(cart) > 0:
                show_cart()
                print("Thank you for shopping with us!")
                shopping = False
            else:
                print("Your cart is empty. Please add items to your cart before checking out.")

        elif choice == '5':
            shopping = False

        else:
            print("Invalid choice. Please choose a valid option (1-5).")

if __name__ == "__main__":
    main()
```

```python
def food_menu():
    print("╔════════════════════════════╗")
    print("║       MENU AND PRICES      ║")
    print("╟────────────────────────────╢")
    for i in range(len(items)):
        print(f"║ {i + 1}. {items[i]:<13} - ${price[i]:.2f}  ║")
    print("╚════════════════════════════╝")

def available_option():
    print("╔════════════════════════════╗")
    print("║      SHOPPING OPTIONS      ║")
    print("╟────────────────────────────╢")
    print("║ 1. View Menu               ║")
    print("║ 2. Add to Cart             ║")
    print("║ 3. View Cart               ║")
    print("║ 4. Checkout                ║")
    print("║ 5. Exit                    ║")
    print("╚════════════════════════════╝")

# Add Data into Cart
def add_to_cart():
    select = int(input("Select Item to Add into Cart: "))
    if 1 <= select <= len(items):
        print(f"Adding {items[select - 1]} to your cart.")
        add_to_cart_product.append(items[select - 1])
        add_to_cart_price.append(price[select - 1])
        while True:
            quantity = int(input("Enter Quantity (1-10): "))
            if 1 <= quantity <= 10:
                qty.append(quantity)
                break
            else:
                print("Invalid quantity. Please choose a quantity between 1 and 10.")
    more_shopping = input("Do you want to add more products ('Y/N'): ").upper()
    if more_shopping == 'Y':
        food_menu()
        add_to_cart()

def view_cart():
    if not add_to_cart_product:
        print("Your cart is empty.")
    else:
        print("╔═══════════════════════════════════╗")
        print("║         YOUR CART                 ║")
        print("╟───────────────────────────────────╢")
        total_cost = 0
        for i in range(len(add_to_cart_product)):
            item, price, quantity = add_to_cart_product[i], add_to_cart_price[i], qty[i]
            total_cost += price * quantity
            print(f"║ {i + 1}. {item:<15} ${price:.2f} x {quantity:<5} ║")
        print("╟───────────────────────────────────╢")
        print(f"║ Total: ${total_cost:<13.2f}            ║")
        print("╚═══════════════════════════════════╝")


def check_out():
    view_cart()
    total_price()

def total_price():
    if add_to_cart_product:
        sum = 0
        for i in range(len(add_to_cart_price)):
            sum = int(sum) + int(add_to_cart_price[i] * qty[i])
        print(f"Your Total bill is: ${sum:.2f}")

price = [10, 20, 30]
items = ["Coffee", "Tea", "Green Tea"]
add_to_cart_product = []
add_to_cart_price = []
qty = []

print("WELCOME TO ONLINE FOOD STORE")

while True:
    available_option()
    choice = int(input("Enter Your Choice: "))
    if choice == 1:
        food_menu()
    elif choice == 2:
        food_menu()
        add_to_cart()
    elif choice == 3:
        view_cart()
    elif choice == 4:
        check_out()
    elif choice == 5:
        print("Thank you for choosing us. See you soon!")
        break
    else:
        print("Invalid choice. Please select a valid option (1-5).")
```
