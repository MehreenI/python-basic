```python
i = 1
while i<=5:
    print(" " * (5-i) ,end="")
    j=1
    while j <= i:
        print("*",end="")
        j=j+1
        
    i = i + 1
    print()
    
print("Done")
```

```python
i = 1
while i <= 5:
    # Print spaces before the asterisks
    print(" " * (5 - i), end="")
    
    j = 1
    while j <= i:
        # Print the asterisks
        print("*", end="")
        j = j + 1

    # Move to the next line for the next row
    print()
    i = i + 1

print("Done")
```

# GUESS THE NUMBER
#### import random library 
#### random.randint(initial_limit,final_limit)
#### random_num variable --> store Random Numbers
#### variable to store score --> inital == 100
#### attempts variable to store no of attempts ---> initail == 0

#### flag variable = true Game is start || falg == False game is finish

#### Ask name from user: 
#### heading Guess the number game 
#### guess variable ---> to store user guessed number
#### if guess == random_num 
#### print you Guessed it 
#### print score no of attemps and set flag to false
#### if guess < random_num 
#### print Number is greater than {guess} number ----> atemptt++ score-= 10

#### if guess > random_num 
#### print Number is less than {guess} number ----> atemptt++ score-= 10


```python
import random
random_num = random.randint(0,50)
score = 100
attempt = 0
flag = True

name = input("Enter Your Name: ").upper()
print("""
 _____                                                                                                      _____ 
( ___ )----------------------------------------------------------------------------------------------------( ___ )
 |   |                                                                                                      |   | 
 |   |                                                                                                      |   | 
 |   |                                                                                                      |   | 
 |   |     ____                          _____  _             _   _                    _                    |   | 
 |   |    / ___| _   _   ___  ___  ___  |_   _|| |__    ___  | \ | | _   _  _ __ ___  | |__    ___  _ __    |   | 
 |   |   | |  _ | | | | / _ \/ __|/ __|   | |  | '_ \  / _ \ |  \| || | | || '_ ` _ \ | '_ \  / _ \| '__|   |   | 
 |   |   | |_| || |_| ||  __/\__ \\__ \   | |  | | | ||  __/ | |\  || |_| || | | | | || |_) ||  __/| |      |   | 
 |   |    \____| \__,_| \___||___/|___/   |_|  |_| |_| \___| |_| \_| \__,_||_| |_| |_||_.__/  \___||_|      |   | 
 |   |     ____                                                                                             |   | 
 |   |    / ___|  __ _  _ __ ___    ___                                                                     |   | 
 |   |   | |  _  / _` || '_ ` _ \  / _ \                                                                    |   | 
 |   |   | |_| || (_| || | | | | ||  __/                                                                    |   | 
 |   |    \____| \__,_||_| |_| |_| \___|                                                                    |   | 
 |   |                                                                                                      |   | 
 |   |                                                                                                      |   | 
 |___|                                                                                                      |___| 
(_____)----------------------------------------------------------------------------------------------------(_____)
""")
while flag:
    guess = int(input("Enter Guessed Number: "))
    print("════════════════════════════════════════════════════════════════════")
    print()
    if(guess == random_num):
        print("Very Well Played, {}".format(name))
        print("No of attempts used = "+str(attempt))
        print("Your Score is = " +str(score))
        flag = False
        
    elif guess < random_num:
        print ( f"Actual Number is Greater Than {guess} number\n\n")
        attempt = attempt + 1;
        score = score - 10
        
    elif guess > random_num:
        print ( f"Actual Number is Less Than {guess} number\n\n")
        attempt = attempt + 1;
        score = score - 10
    
```
